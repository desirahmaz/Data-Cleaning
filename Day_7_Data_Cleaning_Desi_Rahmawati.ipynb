#3. Rename Kolom

```python
#Rename column customerID menjadi CustID
df=df.rename(columns={'customerID':'CustID'})
```

```python
#Mengubah semua column menjadi lowercase
df.columns=df.columns.str.lower()
```

#4. Data Cleaning: Hilangkan "Automatic" pada Payment Method

```python
#Menghilangkan kata automatic pada kolom Payment Method

df['paymentmethod']=df['paymentmethod'].str.replace('(automatic)','')
```

```python
df['paymentmethod']
```

```python
#Menghilangkan tanda '()' pada payment method

df['paymentmethod']=df['paymentmethod'].str.replace('()','',regex=False)
```

```python
#Melakukan pengecekan kembali hasil command
df['paymentmethod']
```

```python
#menghilangkan karakter spasi diawal dan diakhir kalimat pada payment method (menggunakan fungsi strip)
df['paymentmethod']=df['paymentmethod'].str.strip()
```

```python
df
```

resul: index 488 data bukan berupa number (float) namun berupa spasi

```python
#mengecek index berapa saja yang bukan float. True=index yang bersifat float, False=bukan float

def isnumber(x):
  try:
    float(x)
    return True
  except:
    return False

df['isnumber']=df['totalcharges'].apply(isnumber)
index_isnotnumber=df[~df['isnumber']].index
```

```python
#mengecek index yang isinya False (data bukan float) ada di index mana saja
df[~df['isnumber']]
```

```python
#fill totalcharges not number dengan median
#cari median totalcharges
median=df[df['isnumber']]['totalcharges'].astype('float64').median()
```

```python
#membaca hasil median
median
```

```python
#mengubah 11 data pada variabel totalcharges dengan menggunakan median
df.loc[index_isnotnumber,'totalcharges']= median
```

```python
#mengecek kembali apakah sudah berganti menjadi float
df.loc[index_isnotnumber].dtypes
```

```python
#hasil diatas ternyata variable totalcharges belum berubah menjadi float
#maka,
#memgubah variable totalcharges menjadi data float
df['totalcharges']=df['totalcharges'].astype('float64')
```
